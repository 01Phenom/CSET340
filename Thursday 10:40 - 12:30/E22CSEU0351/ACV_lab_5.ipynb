```python
import cv2
from PIL import Image

def compress_image_lossy(input_path, output_path, quality=50):
    # Read the image using OpenCV
    image = cv2.imread(input_path)
    # Save as JPEG with the specified quality
    cv2.imwrite(output_path, image, [int(cv2.IMWRITE_JPEG_QUALITY), quality])
    print(f"Lossy compression (JPEG) done: {output_path}")

def compress_image_lossless(input_path, output_path):
    # Open the image using PIL
    image = Image.open(input_path)
    # Save as PNG (lossless compression)
    image.save(output_path, format='PNG', optimize=True)
    print(f"Lossless compression (PNG) done: {output_path}")

# Example usage
input_image = "/content/po.jpeg"  # Change to your image path
output_jpeg = "compressed_lossy.jpg"
output_png = "compressed_lossless.png"

compress_image_lossy(input_image, output_jpeg)
compress_image_lossless(input_image, output_png)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.datasets import mnist, cifar10
from sklearn.model_selection import train_test_split

# Function to create a CNN model
def create_cnn(input_shape, num_classes):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Function to evaluate the model with ROC and AUC
def evaluate_model_with_roc_auc(model, X_test, y_test, num_classes):
    print("Evaluating the model...")
    y_pred_prob = model.predict(X_test)
    y_pred = np.argmax(y_pred_prob, axis=1)
    y_test_true = np.argmax(y_test, axis=1)

    # Classification Report & Confusion Matrix
    print("Classification Report:")
    print(classification_report(y_test_true, y_pred))

    cm = confusion_matrix(y_test_true, y_pred)
    print("Confusion Matrix:\n", cm)

    # Compute ROC curve and AUC for each class
    fpr, tpr, roc_auc = {}, {}, {}
    for i in range(num_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_pred_prob[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Compute micro-average ROC curve and AUC
    fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_pred_prob.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    # Plot ROC curve
    plt.figure()
    plt.plot(fpr["micro"], tpr["micro"], label=f"Micro-average ROC curve (AUC = {roc_auc['micro']:.2f})")
    for i in range(num_classes):
        plt.plot(fpr[i], tpr[i], label=f"Class {i} (AUC = {roc_auc[i]:.2f})")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend(loc="lower right")
    plt.show()

    return cm, roc_auc

# MNIST Dataset Processing
print("Loading MNIST dataset...")
(mnist_X_train, mnist_y_train), (mnist_X_test, mnist_y_test) = mnist.load_data()
mnist_X_train, mnist_X_test = mnist_X_train / 255.0, mnist_X_test / 255.0
mnist_X_train = mnist_X_train.reshape(-1, 28, 28, 1)
mnist_X_test = mnist_X_test.reshape(-1, 28, 28, 1)
mnist_y_train = to_categorical(mnist_y_train, 10)
mnist_y_test = to_categorical(mnist_y_test, 10)
X_train_mnist, X_val_mnist, y_train_mnist, y_val_mnist = train_test_split(mnist_X_train, mnist_y_train, test_size=0.2, random_state=42)

# Train CNN on MNIST
print("Training CNN on MNIST dataset...")
mnist_model = create_cnn((28, 28, 1), 10)
mnist_model.fit(X_train_mnist, y_train_mnist, epochs=10, validation_data=(X_val_mnist, y_val_mnist))

# Evaluate model on MNIST
evaluate_model_with_roc_auc(mnist_model, mnist_X_test, mnist_y_test, 10)

# CIFAR-10 Dataset Processing
print("Loading CIFAR-10 dataset...")
(cifar_X_train, cifar_y_train), (cifar_X_test, cifar_y_test) = cifar10.load_data()
cifar_X_train, cifar_X_test = cifar_X_train / 255.0, cifar_X_test / 255.0
cifar_y_train = to_categorical(cifar_y_train, 10)
cifar_y_test = to_categorical(cifar_y_test, 10)
X_train_cifar, X_val_cifar, y_train_cifar, y_val_cifar = train_test_split(cifar_X_train, cifar_y_train, test_size=0.2, random_state=42)

# Train CNN on CIFAR-10
print("Training CNN on CIFAR-10 dataset...")
cifar_model = create_cnn((32, 32, 3), 10)
cifar_model.fit(X_train_cifar, y_train_cifar, epochs=10, validation_data=(X_val_cifar, y_val_cifar))

# Evaluate model on CIFAR-10
evaluate_model_with_roc_auc(cifar_model, cifar_X_test, cifar_y_test, 10)
```
