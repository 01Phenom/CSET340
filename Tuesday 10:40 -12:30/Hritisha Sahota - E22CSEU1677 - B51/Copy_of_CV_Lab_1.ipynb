```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

image = cv2.imread('/content/Image.jpeg')
```

```python
def translate_image(image, tx, ty):
    rows, cols = image.shape[:2]
    M = np.float32([[1, 0, tx], [0, 1, ty]])
    translated_image = cv2.warpAffine(image, M, (cols, rows))
    return translated_image
```

```python
def reflect_image(image):
    reflected_image = cv2.flip(image, 1)
    return reflected_image
```

```python
def rotate_image(image, angle):
    rows, cols = image.shape[:2]
    M = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
    rotated_image = cv2.warpAffine(image, M, (cols, rows))
    return rotated_image
```

```python
def scale_image(image, scale_x, scale_y):
    scaled_image = cv2.resize(image, None, fx=scale_x, fy=scale_y, interpolation=cv2.INTER_LINEAR)
    return scaled_image
```

```python
def shear_x(image, shear_factor):
    rows, cols = image.shape[:2]
    M = np.float32([[1, shear_factor, 0], [0, 1, 0]])
    sheared_image = cv2.warpAffine(image, M, (cols, rows))
    return sheared_image
```

```python
def shear_y(image, shear_factor):
    rows, cols = image.shape[:2]
    M = np.float32([[1, 0, 0], [shear_factor, 1, 0]])
    sheared_image = cv2.warpAffine(image, M, (cols, rows))
    return sheared_image
```

```python
def crop_image(image, start_row, start_col, end_row, end_col):
    cropped_image = image[start_row:end_row, start_col:end_col]
    return cropped_image
```

```python
translated = translate_image(image, 50, 30)
reflected = reflect_image(image)
rotated = rotate_image(image, 45)
scaled = scale_image(image, 1.5, 1.5)
cropped = crop_image(image, 50, 50, 200, 200)
sheared_x = shear_x(image, 0.5)
sheared_y = shear_y(image, 0.5)
```

```python
cv2_imshow(translated)
cv2_imshow(reflected)
cv2_imshow(rotated)
cv2_imshow(scaled)
cv2_imshow(cropped)
cv2_imshow(sheared_x)
cv2_imshow(sheared_y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def translate(points, tx, ty):
    translation_matrix = np.array([[1, 0, tx],
                                    [0, 1, ty],
                                    [0, 0, 1]])
    return np.dot(points, translation_matrix.T)

def scale(points, sx, sy):
    scaling_matrix = np.array([[sx, 0, 0],
                                [0, sy, 0],
                                [0, 0, 1]])
    return np.dot(points, scaling_matrix.T)

def rotate(points, angle):
    radians = np.radians(angle)
    rotation_matrix = np.array([[np.cos(radians), -np.sin(radians), 0],
                                 [np.sin(radians), np.cos(radians), 0],
                                 [0, 0, 1]])
    return np.dot(points, rotation_matrix.T)

def reflect(points):
    reflection_matrix = np.array([[1, 0, 0],
                                   [0, -1, 0],
                                   [0, 0, 1]])
    return np.dot(points, reflection_matrix.T)

def shear(points, shx, shy):
    shearing_matrix = np.array([[1, shx, 0],
                                 [shy, 1, 0],
                                 [0, 0, 1]])
    return np.dot(points, shearing_matrix.T)
```

```python
def plot_shapes(original, transformed_shapes, titles):
    plt.figure(figsize=(12, 8))

    plt.subplot(2, 3, 1)
    plt.plot(original[:, 0], original[:, 1], 'b-o')
    plt.title('Original Shape')
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.axhline(0, color='black',linewidth=0.5, ls='--')
    plt.axvline(0, color='black',linewidth=0.5, ls='--')

    for i, shape in enumerate(transformed_shapes):
        plt.subplot(2, 3, i + 2)
        plt.plot(shape[:, 0], shape[:, 1], 'r-o')
        plt.title(titles[i])
        plt.xlim(-10, 10)
        plt.ylim(-10, 10)
        plt.axhline(0, color='black',linewidth=0.5, ls='--')
        plt.axvline(0, color='black',linewidth=0.5, ls='--')

    plt.tight_layout()
    plt.show()
```

```python
original_shape = np.array([[1, 1, 1],
                            [2, 1, 1],
                            [2, 2, 1],
                            [1, 2, 1],
                            [1, 1, 1]])
```

```python

translated_shape = translate(original_shape, 2, 1)
scaled_shape = scale(original_shape, 1.5, 1.5)
rotated_shape = rotate(original_shape, 45)
reflected_shape = reflect(original_shape)
sheared_shape = shear(original_shape, 0.5, 0)


transformed_shapes = [translated_shape, scaled_shape, rotated_shape, reflected_shape, sheared_shape]
titles = ['Translated', 'Scaled', 'Rotated', 'Reflected', 'Sheared']
```

```python
plot_shapes(original_shape, transformed_shapes, titles)
```

```python
import numpy as np
import matplotlib.pyplot as plt


def translate(points, tx, ty):
    translation_matrix = np.array([[1, 0, tx],
                                    [0, 1, ty],
                                    [0, 0, 1]])
    return np.dot(points, translation_matrix.T)

def scale(points, sx, sy):
    scaling_matrix = np.array([[sx, 0, 0],
                                [0, sy, 0],
                                [0, 0, 1]])
    return np.dot(points, scaling_matrix.T)

def rotate(points, angle):
    radians = np.radians(angle)
    rotation_matrix = np.array([[np.cos(radians), -np.sin(radians), 0],
                                 [np.sin(radians), np.cos(radians), 0],
                                 [0, 0, 1]])
    return np.dot(points, rotation_matrix.T)

def reflect(points):
    reflection_matrix = np.array([[1, 0, 0],
                                   [0, -1, 0],
                                   [0, 0, 1]])
    return np.dot(points, reflection_matrix.T)

def shear(points, shx, shy):
    shearing_matrix = np.array([[1, shx, 0],
                                 [shy, 1, 0],
                                 [0, 0, 1]])
    return np.dot(points, shearing_matrix.T)


original_shape = np.array([[1, 1, 1],
                            [2, 1, 1],
                            [2, 2, 1],
                            [1, 2, 1],
                            [1, 1, 1]])
translated_shape = translate(original_shape, 2, 1)
scaled_shape = scale(original_shape, 1.5, 1.5)
rotated_shape = rotate(original_shape, 45)
reflected_shape = reflect(original_shape)
sheared_shape = shear(original_shape, 0.5, 0)

plt.figure(figsize=(10, 8))

plt.plot(original_shape[:, 0], original_shape[:, 1], 'b-o', label='Original Shape')

plt.plot(translated_shape[:, 0], translated_shape[:, 1], 'r-o', label='Translated')
plt.plot(scaled_shape[:, 0], scaled_shape[:, 1], 'g-o', label='Scaled')
plt.plot(rotated_shape[:, 0], rotated_shape[:, 1], 'm-o', label='Rotated')
plt.plot(reflected_shape[:, 0], reflected_shape[:, 1], 'c-o', label='Reflected')
plt.plot(sheared_shape[:, 0], sheared_shape[:, 1], 'y-o', label='Sheared')

plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.title('2D Transformations on a Shape')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.legend()
plt.show()
```
